```python
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load the California housing dataset
california = fetch_california_housing()

# Display feature names and a preview of the dataset
print("Feature names:", california['feature_names'])
print("First five rows of data:\n", california['data'][:5])

# Use only one feature for simplicity; let's choose 'MedInc' (median income in block group)
x = california.data[:, np.newaxis, 0]  # 'MedInc' is at index 0
y = california.target  # Define the target variable (median house value in $100,000s)

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)

# Create a Linear Regression model
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# Make predictions on the training and testing sets
y_train_pred = regressor.predict(x_train)
y_test_pred = regressor.predict(x_test)

# Plot the training data and the model's prediction line
mtp.scatter(x_train, y_train, color="green")
mtp.plot(x_train, y_train_pred, color="red")
mtp.title("Linear Regression on California Housing Dataset (Training Set)")
mtp.xlabel("Median Income in Block Group (MedInc)")
mtp.ylabel("Median House Value in $100,000s")
mtp.show()

# Plot the testing data and the model's prediction line
mtp.scatter(x_test, y_test, color="blue")
mtp.plot(x_train, y_train_pred, color="red")  # Use training line as reference
mtp.title("Linear Regression on California Housing Dataset (Testing Set)")
mtp.xlabel("Median Income in Block Group (MedInc)")
mtp.ylabel("Median House Value in $100,000s")
mtp.show()

# Output the coefficients and intercept of the model
print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)

# Calculate and print the mean squared error and R^2 score
print('Mean Squared Error:', mean_squared_error(y_test, y_test_pred))
print('Coefficient of Determination (R^2 Score):', r2_score(y_test, y_test_pred))
```
